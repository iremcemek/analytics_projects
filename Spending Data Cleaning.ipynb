```python
import pandas as pd
import os
import time
from pyspark.sql.functions import col

# 1. ADIM: Veriyi Spark ile Oku
table_name = "hive_metastore.dev_country_turkey.spending_joined_4_txt"
df = spark.table(table_name)

print(f"Tablo Okundu. Satır Sayısı: {df.count()}")

# 2. ADIM: Pandas'a Çevir (5555 satır için en güvenli yöntem)
pdf = df.toPandas()

# --- ÖNEMLİ KONTROL: Veri Tipi ---
# Bazen Excel'den gelen verilerde 707 sayısı "707" (metin) olarak gelebilir.
# Bunu garantiye almak için _c11 sütununu sayıya çeviriyoruz.
pdf['_c11'] = pd.to_numeric(pdf['_c11'], errors='coerce')

# 3. ADIM: Çift Koşullu Değişiklik
# Koşul: _c12 == 'DKESEL01' VE _c11 == 707 ise -> 755 yap
condition = (pdf['_c12'] == 'DKESEL01') & (pdf['_c11'] == 707)

# Etkilenen satır sayısını görelim
affected_rows = pdf[condition].shape[0]
print(f"Koşula uyan ve değiştirilecek satır sayısı: {affected_rows}")

# Değişikliği uygula
pdf.loc[condition, '_c11'] = 755

# 4. ADIM: Dosyayı Yazma (Cache Sorununu Önlemek İçin Benzersiz İsim)
# Her seferinde farklı bir isim veriyoruz ki tarayıcı eski dosyayı getirmesin.
ts = int(time.time())
file_name = f"spending_final_{ts}.csv"
file_path = f"/dbfs/FileStore/tables/{file_name}"

# Eski dosyaları temizlemeye gerek yok, yeni isim kullanıyoruz.
pdf.to_csv(file_path, index=False, header=False)

# 5. ADIM: İndirme Linki
# Dosya boyutunu kontrol edelim
if os.path.exists(file_path):
    size = os.path.getsize(file_path)
    print(f"✅ Dosya oluşturuldu: {file_name}")
    print(f"📂 Dosya Boyutu: {size} bytes")
    
    displayHTML(f"""
    <div style="border:2px solid green; padding: 10px; background-color: #f0fff0;">
        <h3>✅ İşlem Başarılı!</h3>
        <p><b>{affected_rows}</b> adet satır güncellendi.</p>
        <p>Dosya ID: {ts}</p>
        <a href='/files/tables/{file_name}' download style="font-size:16px; font-weight:bold; color:blue;">
           👉 Buraya Tıklayarak İndir: {file_name}
        </a>
    </div>
    """)
else:
    print("❌ HATA: Dosya yazma işlemi başarısız oldu.")
```
